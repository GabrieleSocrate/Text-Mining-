```python
import pandas as pd

#Percorso dataset
file_path = "/content/drive/MyDrive/Ds/financial_news_base.jsonl"

df = pd.read_json(file_path, lines=True)

df.head()
```

```python
from sklearn.model_selection import train_test_split

#Riduzioni osservazioni
df_small, _ = train_test_split(
    df,
    train_size=25000,
    stratify=df['sentiment_label'],
    random_state=100
)

# Divisione 70/30
train_df, temp_df = train_test_split(
    df_small,
    test_size=0.30,
    stratify=df_small['sentiment_label'],
    random_state=100
)

# Divisione Val e Test
val_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    stratify=temp_df['sentiment_label'],
    random_state=100
)

print(len(train_df))
print(len(val_df))
print(len(test_df))

def print_distribution(data, name):
    counts = data['sentiment_label'].value_counts()
    perc = data['sentiment_label'].value_counts(normalize=True) * 100
    dist = pd.DataFrame({
        'count': counts,
        'percentage (%)': perc.round(2)
    })
    print(f"\n{name}")
    print(dist)

print_distribution(df, "Full dataset")
print_distribution(train_df, "Train set")
print_distribution(val_df, "Validation set")
print_distribution(test_df, "Test set")
```
